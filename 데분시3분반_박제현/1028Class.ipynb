```python
#평균을 산출할 수 있는 레이블만 요약해준다.
#크기가 상관없는 변수: 범주형
df.describe(include = 'all')
```

```python
#astype은 enplace 매겨변수를 쓸 수 없다.
df = df.astype({'survived': 'category'})
```

```python
df['survived'].describe()
```

```python
vis1 = df['survived'].value_counts()
vis1.plot.bar(rot = 0)
```

```python
vis1.index = ['No', 'Yes']
vis1
vis1.plot.bar(rot = 0)
```

```python
sv = df['survived'].value_counts(normalize = True) * 100
sv.index = ['No', 'Yes']
sv.sort_index(ascending = False, inplace = True)
sv.plot.bar(rot = 0)
```
